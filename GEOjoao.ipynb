• Crie um mapa usando a biblioteca “folium” e localize as seguintes
cidades:
• # Coordenadas das cidades (normalmente, você obteria essas coordenadas
de uma base de dados ou API). Aqui está a geolocalizacao delas:
coordenadas = {
'Porto Alegre': [-30.0346, -51.2177],
'São Paulo': [-23.5505, -46.6333],
'Rio de Janeiro': [-22.9068, -43.1729],
'Curitiba': [-25.4290, -49.2671],
'Belo Horizonte': [-19.9173, -43.9345]
}

```python

import folium


#Definindo coordenadas Porto Alegre
#latitude = -30.0346
#longitude = -51.2177

lat = [-30.0346, -23.5505, -22.9068,-25.4290,-19.9173]  

long = [-51.2177,-46.6333, -43.1729, -49.2671, -43.9345]  

cidade = ['Porto Alegre', 'São Paulo', 'Rio de Janeiro', 'Curitiba', 'Belo Horizonte']

#criando mapa

mapa = folium.Map(location=[lat[0], long[0]], zoom_start=12)

#add marcador 
for i in range(5):
    
    folium.Marker([lat[i], long[i]], popup = cidade[i]).add_to(mapa)

mapa
```
